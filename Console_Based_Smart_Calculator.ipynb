## Problem Statement

The aim of this project is to create a console-based smart calculator using Python and NumPy. The calculator performs multiple
mathematical operations using Python and NumPy like addition, subtraction, multiplication, division, power, modulus, and NumPy array statistics.

## Logic Explanation

1.Display menu options.
2.Take user choice.
3.Ask for numbers.
4.Perform calculation.
5.Display result.
6.Ask whether user wants to continue or exit.

```python
import numpy as np

def show_menu():
  print("\n**************************************")
  print("       SMART CALCULATOR MENU")
  print("**************************************")
  print("1.Addition")
  print("2.Subtraction")
  print("3.Multiplication")
  print("4.Division")
  print("5.Power")
  print("6.Modulus")
  print("7.NumPy Array Operations")
  print("8.Exit")
  print("**************************************")


def get_count():
  count=input("How many numbers?")
  while count.isdigit()==False or int(count)<=0:
    print("Please enter whole numbers only")
    count=input("How many numbers?")
  return int(count)


def addition():
  n=get_count()
  numbers=[]
  i=0
  while i<n:
    num=float(input("Enter number:"))
    numbers.append(num)
    i=i+1
  arr=np.array(numbers)
  result=np.sum(arr)
  print("Numbers:",numbers)
  print("Sum=",result)


def subtraction():
  a=float(input("Enter first number:"))
  b=float(input("Enter second number:"))
  result=a-b
  print("Subtraction result=",result)


def multiplication():
  n=get_count()
  numbers=[]
  i=0
  while i<n:
    num=float(input("Enter number:"))
    numbers.append(num)
    i=i+1
  result=1
  for num in numbers:
    result=result*num
  print("Numbers:",numbers)
  print("Product=",result)


def division():
  a=float(input("Enter dividend:"))
  b=float(input("Enter divisor:"))
  if b==0:
    print("Error: Division by zero is not allowed")
  else:
    result=a/b
    print("Division result=",result)


def power():
  a=float(input("Enter base:"))
  b=float(input("Enter exponent:"))
  result=a**b
  print("Power result=",result)


def modulus():
  a=float(input("Enter dividend:"))
  b=float(input("Enter divisor:"))
  if b==0:
    print("Error: Division by zero is not allowed")
  else:
    result=a%b
    print("Modulus result=",result)


def numpy_stats():
  n=get_count()
  numbers=[]
  i=0
  while i<n:
    num=float(input("Enter number:"))
    numbers.append(num)
    i=i+1
  arr=np.array(numbers)
  print("Array:",arr)
  print("Sum=",np.sum(arr))
  print("Mean=",np.mean(arr))
  print("Median=",np.median(arr))
  print("Standard Deviation=",np.std(arr))
  print("Variance=",np.var(arr))



print("Welcome to Smart Calculator!")
choice="0"
while choice!="8":
  show_menu()
  choice=input("Enter your choice (1-8):")
  if choice=="1":
    addition()
  elif choice=="2":
    subtraction()
  elif choice=="3":
    multiplication()
  elif choice=="4":
    division()
  elif choice=="5":
    power()
  elif choice=="6":
    modulus()
  elif choice=="7":
    numpy_stats()
  elif choice=="8":
    print("Thank you! Goodbye")
  else:
    print("Invalid choice.Please enter 1 to 8.")
  if choice!="8":
    again=input("Continue? (yes/no): ")
    if again=="no":
      print("Thank you! Goodbye")
      choice="8"
```
